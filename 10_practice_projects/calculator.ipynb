```python
# simple calculator

# 3 step

# functions for operations
# user input
# print result

# function to add
def add(num1, num2):
    return num1 + num2

# function to subtract
def sub(num1, num2):
    return num1 - num2

# function to multiply
def prod(num1, num2):
    return num1 * num2

# divide two numbers
def div(num1, num2):
    return num1 / num2

# mean of two numbers
def mean(num1, num2):
    return (num1 + num2)/2

# user input

print("please select a operation: \n  " \
     "1. Addition\n" \
     "2. Substraction\n" \
     "3. Multiplication\n" \
     "4. Division\n" \
     "5. Average\n")

select = int(input("Select a operation from 1,2,3,4,5: "))

number1 = int(input("Enter first number: "))
number2 = int(input("Enter second number: "))

if select == 1:
    print(f'Sum of {number1} and {number2} is: {add(number1, number2)}')

elif select == 2:
    print(f'Sub of {number1} and {number2} is: {sub(number1, number2)}')

elif select == 3:
    print(f'Multiplication of {number1} and {number2} is: {prod(number1, number2)}')
    
elif select == 4:
    print(f'division of {number1} and {number2} is: {div(number1, number2)}')
    
elif select == 5:
    print(f'average of {number1} and {number2} is: {mean(number1, number2)}')
else:
    print("Invalid input")

```
